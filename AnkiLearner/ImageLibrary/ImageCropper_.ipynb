# Image cropper

```python
import os
import cv2
from os import listdir
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def cropImage(images, t,b,l,r):
    croppedImages = []
    # find the cropped coordinations
    for i in range (len(images)):
        h , w,_  = images[i].shape
        xm = w//2
        ym = h//2
        x0 = xm - l
        x1 = xm + r
        y0 = ym - t
        y1 = ym + b
        # crop the image
        croppedImg = images[i][y0:y1, x0:x1]
        croppedImages.append(croppedImg)
    return croppedImages
```

```python
folders = listdir()
print (os.getcwd())
print (folders)
for path in folders:
    if os.path.isdir(path):
        if path[0] != ".":
            print (path)
            files = (listdir(path))
            names = [os.path.join(path,f) for f in files if f.split(".")[-1] == "jpg"]
            print ("#: ", len(names))
            imgs = [cv2.imread(name) for name in names ]
            cropped = cropImage(imgs,100,100,125,125)
            for i in range (len(names)):
                print (i, names[i], len(cropped))
                cv2.imwrite(names[i], cropped[i])
```
